```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset
credit_card=pd.read_csv('/content/creditcard.csv')
```

```python
#data processing
credit_card.head()
```

```python
credit_card.shape
```

```python
credit_card.info()
```

```python
credit_card.describe()
```

```python
credit_card.isnull().sum()
```

```python
credit_card=credit_card.dropna()
```

```python
credit_card.isnull().sum()
```

```python
credit_card['Class'].value_counts()

#0->>normal Transactions
#1->>1 fradulent Transactions
```

```python
#seperating the data for analysis
legit=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card.groupby('Class').mean()
```

```python
#under-sampling
#build a sample dataset containing similar distribution of normal transactions  and fradulent transactins
legit_sample=legit.sample(n=294)
```

```python
#concatenating 2 dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data into features and targets
x=new_dataset.drop(columns=['Class'],axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#split the data into trining data and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
print(x_train.shape,x_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

```python
#training the model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#finding the accuacy witth testing data
testing_accuracy=model.predict(x_train)
accuracy=accuracy_score(testing_accuracy,y_train)
print(accuracy)
```
